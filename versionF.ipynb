On convertit les poids en kilos  

```python
def convert(i):
    weight = ""
    unite = ""
    for j in i:
        try :
            t = int(j) #ON TESTE JUSTE SI INT J RISE UNE ERREUR 
            weight += j
        except ValueError:
            unite += j

    unite = unite.strip() # on retire les espaces
    if unite == 'lbs':
        weight = int(weight)/2.205 
        
        return weight
    elif unite == 'kg':
        return weight 
    else: 
        return False
    
    
noUnite = []
k = 0 
for i in df["Bag.Weight"]:
    if convert(i) is False:
        noUnite.append(k)
    k += 1
noUnite

df.drop(noUnite,axis=0, inplace = True)

    
```

# The choice of the right country

```python
critical_variables = ['Aroma', 'Flavor','Acidity' ,'Uniformity','Total.Cup.Points']

def RowToBeRm(df , col):
    k = 0
    clean = []
    for i in df[col]:
        if i != i:
            clean.append(k)
        k += 1
    return clean
    
def clean_crit(df, l = critical_variables ):
    clean = []
    for i in l : 
        clean += RowToBeRm(df,i)
    return clean

def one_country(d,name):
    clean = []
    k = 0
    for i in d["Country.of.Origin"]:
        if i != name:
            clean.append(k)
        k += 1
    clean += clean_crit(d, critical_variables)
    clean = list(set(clean))
    d.drop(clean, axis =0 , inplace = True)



def wanted_criteria (df, wanted_value) : 
    """takes in parameters a dataframe and a tuple which contains a columns and the value one wants to find in it """
    df[wanted_value[0]].astype('float')
    gap = 10
    for i in df[wanted_value[0]].astype('float'):
        if abs(wanted_value[1] - i)<gap:
            gap = abs(wanted_value[1] - i)
    print (wanted_value[0] + " gap with the wanted value is" , gap)
    return gap

def wanted_criteria_frequency (df, wanted_value, precision) : 
    """takes in parameters a dataframe and a tuple which contains a columns and the value one wants to find in it """
    match = 0
    for i in df[wanted_value[0]].astype('float'):
        if abs(wanted_value[1] - i)<precision:
            match += 1
    if match<2 : 
        plural = "producer of coffee"
    else : 
        plural = "producers of coffee"
    print (wanted_value[0] + " gap with the wanted value is less than " , precision," for ", match,plural)
    return match
```

```python
l = [brazil,burundi,ethiopia,honduras, nicaragua]
l_best_producteur = ['Brazil','Burundi', 'Ethiopia', 'Honduras', 'Nicaragua']
wanted_values = [('Aroma',wanted_aroma),('Flavor',wanted_flavor), ('Uniformity', 10),('Acidity', wanted_acidity) ]
k = 0
for i in l :
    
    one_country(i, l_best_producteur[k]) #o ne garde que les lignes qui concerne le pays i
    print('\n',l_best_producteur[k])
    gap_mean = 0
    for j in wanted_values :
        wanted_criteria(i,j)
        wanted_criteria_frequency(i,j,0.1)
    k += 1
       
    
```
